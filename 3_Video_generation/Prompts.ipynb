### Пример промпта для генрации сегмента видео

#### Сегмент 23

Промпт генерации видео для модели Veo 3.1:

```
Крупный план: Рука Генерального директора (мужчина, азиатская внешность (Китай/Корея), 50 лет, в безупречном костюме) спокойно берет платиновый брелок с ключами со стола. Он уходит домой.


Персона: Генеральный директор. Мужчина, азиатская внешность (Китай/Корея), 50 лет, в безупречном костюме.

Сцена: Минималистичный и большой современный конференц зал современного небоскрёба. На левой стене весит ЖК дисплей. Зал заканчивается большим панорамным окном. За окном закат. По центру зала распологается длинный стол для проведения переговоров из лакированного дерева.

Действие: Директор стоит в минималистичном зале слева от стола и смотрит на сгенерированный контент на монитор. Лицо выражает глубокое удовлетворение и контроль. Камера приближается к нему, затем останавливается. Директор смотрит в камеру и улыбается.

МАКСИМАЛЬНАЯ ДЕТАЛИЗАЦИЯ, РЕАЛИЗМ
```
