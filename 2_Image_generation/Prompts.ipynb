# Промпт для разработки изображения для закреплённой записи

## Задача

Необходимо создать одно квадратное изображение, которое будет использоваться как обложка закреплённой записи в социальных сетях компании.

Изображение должно:
- отражать суть компании, её услуги, товары, сильные стороны;
- либо демонстрировать процесс оказания услуги;
- либо передавать смысл компании через метафору или визуальное сравнение.

Текст на изображении может быть, но не обязателен.

Требования к оформлению результата:
- Опишите идею, которую хотите передать этим изображением.
- Сформулируйте промпт (текстовый запрос к нейросети).
- В качестве результата необходимо приложить одно лучшее.


## Решение

**Идея изображения:** Изображение демонстрирует ключевой момент высокоуровневого стратегического совещания в штаб-квартире ASTAS. В центре внимания — сотрудничество между нашей компанией и представителями Совета министров африканских стран по водным ресурсам (AMCOW) по решению критической проблемы водоснабжения.

**Локация:** Современный, элегантный конференц-зал ASTAS. Пространство должно быть минималистичным, но технологичным, отражая интеллектуальный характер компании. Большая стеклянная стена с видом на городской пейзаж (возможно, символизирующая глобальный охват) будет уместна, но необязательна, главное — создать ощущение высококлассной корпоративной среды.

**Участники:** За столом сидят шесть человек.

- Сотрудники ASTAS (3 человека):
    - Генеральный директор ASTAS (мужчина, во главе стола): Славянской внешности, в возрасте 40 лет, с короткой стрижкой "полубокс", светлыми волосами с легкой платиновой сединой, в строгом деловом костюме, излучает уверенность, компетентность и сосредоточенность. Его поза должна демонстрировать лидерство и вовлеченность в дискуссию.
    - Ведущий специалист по R&D (мужчина, слева от гендиректора): Славянской внешности, в строгом деловом костюме, сосредоточенно наблюдает за данными на интерактивном столе.
    - Инженер (мужчина, слева от R&D специалиста): Славянской внешности, в строгом деловом костюме, активно взаимодействует с интерактивным дисплеем.
- Представители AMCOW (2 человека справа от гендиректора): Высокопоставленные официальные лица из африканских стран. Их внешний вид — официальный деловой костюм, возможно, с тонкими элементами национальных культурных акцентов (например, узор на галстуке или тонкий орнамент на одежде), что подчеркивает их идентичность без излишнего стереотипизирования. Их выражения лиц должны передавать серьезность, заинтересованность и дипломатичность.

**Технологический элемент (интерактивный стол):** В центре стола находится большой интерактивный дисплей, на котором отображается детальная географическая карта Африки. На карте ярко выделены регионы с водным дефицитом, а также визуализированы потоки данных, графики, аналитические показатели и прогнозы, касающиеся водных ресурсов, инфраструктурных проектов и климатических моделей. Это подчеркивает применение передовых технологий ASTAS в анализе и принятии решений.

**Брендинг ASTAS:** На стене, за директором, на стене, прямо по центру, размещена надпись "ASTAS". Она должен быть хорошо видна, но не доминировать, лишь подтверждая принадлежность места действия.

**Атмосфера:** Общая атмосфера — серьезность, профессионализм, сосредоточенность на решении сложной задачи, а также дух сотрудничества и стратегического мышления. Освещение должно быть сдержанным, но функциональным, подчеркивающим высокотехнологичность пространства.

### Разбор методик в промпте

| Методика | Текст из промпта (где конкретно используется) | Почему это важно |
| :--- | :--- | :--- |
| **Указание контекста и роли** (Context Setting / Role Prompting) | "...high-level strategic meeting in a modern, sleek, and minimalist corporate conference room at ASTAS headquarters." | Задает общую атмосферу и логику происходящего, помогая ИИ понять уместный стиль объектов, интерьера и одежды. |
| **Внешняя ссылка** (Multimodal Reference / Image-to-Text Bridge) | "A sample of the ASTAS logo is provided in the attachment; please use it as a visual reference." | Минимизирует «галлюцинации» ИИ при отрисовке специфических брендовых элементов, обеспечивая визуальную точность логотипа. |
| **Детализация персонажей** (Character Specification / Anthropometry) | "...aged around 40, with distinct Slavic (Russian) facial features. He has short, 'полубокс' style platinum blonde hair..." | Позволяет избежать шаблонных лиц и создает уникальных, узнаваемых персонажей с конкретными этническими и профессиональными чертами. |
| **Композиция и ракурс** (Composition & Camera Framing) | "A wide-angle... image... directly centered behind the head of the table." | Определяет положение «камеры» и симметрию кадра, что критично для создания ощущения масштабности, порядка и официальности. |
| **Световой дизайн и материалы** (Lighting & Material Texture) | "...natural and refined ambient lighting, cinematic quality... large, luminous interactive conference table..." | Свет и текстуры определяют фотореализм. Контраст между свечением стола и мягким общим светом создает глубину и фокус на данных. |
| **Стилистическое уточнение** (Stylistic Cues / Aesthetic Direction) | "...sophisticated, minimalist style... modern, sleek... professional suit... sharp focus." | Направляет ИИ к определенной эстетике «высокого корпоративного стиля», исключая визуальный шум или дешево выглядящие элементы. |
| **Этическое управление** (Cultural Nuance / Bias Mitigation) | "...respectfully indicate cultural background without stereotype... expressions are serious, diplomatic." | Предотвращает создание карикатурных образов, направляя модель в сторону уважительного и профессионального отображения международного сотрудничества. |
| **Отрицательные ограничения** (Negative Constraints / Explicit Exclusion) | "...no text overlays on the image except for the ASTAS text on the wall." | Прямой запрет на лишние артефакты (водяные знаки, бессмысленные буквы на фоне), которые часто возникают при генерации сложных сцен. |

# Коммерческое задание: Генерация серии иллюстраций для бренда

Создать серию изображений, которые можно будет использовать:
- в социальных сетях
- или в печатных буклетах компании.

**Изображения должны:**
- показывать товар или услугу компании
- либо передавать её дух, миссию, ценности, атмосферу, персонал
- либо визуализировать процесс оказания услуги или результат.

**Изображения должны быть именно иллюстрациями, не реалистичными фото.**

**Обязательное условие:** все изображения должны быть выполнены в одном стиле, как будто это серия из одной визуальной фотосессии.


## Решение

> **ПРИМЕЧАНИЕ: Представлено только одно изображение.**

**Название макета:** "Анализ инфраструктурных данных"

Идея изображения заключается в создании цифровой иллюстрации, которая показывает международное сотрудничество в сфере высоких технологий и управления природными ресурсами.

Ключевые элементы:
- **Действующие лица:** Два профессионала. Первый — инженер-миротворец из организации "ASTAS" (мужчина славянской внешности), одетый в современную и строгую корпоративную одежду (поло с надписью "ASTAS"). Второй — молодой местный эксперт (африканской внешности), также одетый в профессиональную инженерную одежду.
- **Сцена:** Они находятся в чистом, высокотехнологичном центре управления крупным водным объектом, например, гидроэлектростанцией. Место действия — солнечный, засушливый регион, что подчеркивает важность управления водными ресурсами.
- **Действие:** Оба специалиста сосредоточенно работают вместе, анализируя сложные данные на огромном экране высокого разрешения. На экране видны графики, карты и диаграммы, связанные с потоками воды, экологическим мониторингом и распределением ресурсов.
- **Фон:** Через большое панорамное окно видна сама гидроэлектростанция, что напрямую связывает их аналитическую работу в помещении с реальным объектом управления.
- **Атмосфера:** Изображение передает дух профессионализма, интеллектуального труда и авторитетности. Основной посыл — мирное научное и инженерное сотрудничество для решения глобальных проблем, таких как нехватка воды, с помощью передовых технологий.

### Разбор методик в промпте

Промпт сочетает жёсткий контроль (детали, ограничения, роли) и мягкое семантическое направление (атмосфера, идеи, сотрудничество). Такая структура хорошо масштабируется: её можно повторно использовать для других регионов, объектов или брендов, меняя лишь отдельные блоки без потери качества результата.

| Методика | Текст из промпта (где используется) | Почему это важно |
|---|---|---|
| Reference-based prompting (опора на референс) | “A sample of the ASTAS logo is provided in the attachment; please use it as a visual reference.” | Фиксирует визуальную идентичность бренда, снижает вариативность и повышает воспроизводимость результата при последующих генерациях. |
| Style anchoring (якорение стиля) | “Photorealistic digital illustration.” | Сразу задаёт уровень реализма и отсекает нежелательные стили (иллюстративный, мультяшный), что критично для корпоративных и инженерных визуалов. |
| Scene decomposition (декомпозиция сцены) | “Two professionals. One man… The second person…” | Разбиение сцены на сущности упрощает интерпретацию модели и повышает стабильность композиции в сложных многообъектных сценах. |
| Role-based prompting (ролевая конкретизация) | “an engineer from ASTAS”, “a representative of the hydroelectric power plant staff” | Явно задаёт социальные и профессиональные роли, что помогает модели корректно отразить взаимодействие и иерархию. |
| Attribute-level detailing (детализация атрибутов) | “clean, modern polo shirt with the ASTAS logo clearly visible and neat professional trousers” | Точная детализация одежды и внешнего вида повышает контроль над визуальными деталями и уменьшает случайные отклонения. |
| Negative constraint prompting (негативные ограничения) | “opaque LCD panel (… not transparent)” | Явно исключает распространённую ошибку модели, заранее направляя генерацию в нужное русло. |
| Functional object description | “displaying complex data visualizations (detailed water flow graphs, environmental impact data, resource distribution maps)” | Объект описывается не только визуально, но и функционально, что усиливает правдоподобие сцены. |
| Environmental context anchoring | “clean, modern control room of a large hydroelectric power plant” | Контекст среды задаёт логику освещения, архитектуры и оборудования, повышая реализм. |
| Geographical & climate hinting | “in a sunny, arid region” | Косвенно влияет на свет, цветовую температуру и атмосферу, не перегружая промпт техническими параметрами. |
| Background layering | “In the background, the hydroelectric power plant is visible through a panoramic window.” | Добавляет глубину сцены и пространственную иерархию, что важно для высококачественных визуализаций. |
| Abstract concept steering (семантическое направление) | “global engineering solutions for critical resources, and cutting-edge technologies” | Направляет модель на смысловой уровень выше конкретных объектов, формируя «идею кадра». |
| Mood & tone prompting | “professional, intellectual, and authoritative” | Управляет невербальными аспектами сцены (позы, выражения, композиция), что сложно задать иначе. |
| Narrative framing | “emphasizing peaceful scientific and engineering cooperation” | Создаёт сюжет и взаимодействие, а не статичную постановку, повышая ценность изображения для презентаций и PR. |
